### Reflection

### 1. Describe your pipeline. As part of the description, explain how you modified the draw_lines() function.

![original image][image0]



My pipeline consisted of the following steps. First, I converted the images to grayscale, then I applied histogram equalization on the images. This was used to detect white lanes on the road by using the inRange() function after the equalization. I convereted the images to hsv to detect the yellow lanes.


![grayscale image after histogram equalization][image1]
<center>grayscale</center>


![image converted to hsv to detect yellow lanes][image2]
<center>hsv</center>


![mask generated for white lanes by using inRange()][image3]
<center>white mask</center>


![mask generated for yellow lanes by using inRange()][image4]
<center>yellow mask</center>


Second, I marked coordinates of the polygon describing the region of interest and used that to extract the white and yellow lanes in the region of interest.


![the polygon showing the region of interest][image5]
<center>ROI</center>


![the final mask after combining both white and yellow lanes inside ROI][image6]
<center>final mask</center>


Third, I applied edge detection on the resulting image to get the edges of the lane lines.


![image after applying edge detection][image7]
<center>edge detection output</center>



Fourth, I applied Probabilistic Hough transform to extract the lines in the image. Then I segregated the points based on the slope of the lines into left lane and right lane points. Then I applied linear regression to get single left lane line and right lane line.


![output after applying linear regression on left lane and right lane points][image8]
<center>Linear regression output</center>


Fifth, The final image is generated as the weighted sum of the lanes generated and original image.


![final lane lines on image][image9]
<center>final output</center>


In order to draw a single line on the left and right lanes, I modified the draw_lines() function by firstly segragating the points into left and right lane points according to the slope of the input points. Then I applied Linear regression seperately on the left and right lane points to get the two lane lines.


### 2. Identify potential shortcomings with your current pipeline


One potential shortcoming would be what would happen when there are erractic shadows or extreme sunlight in the image. 

Another potential shortcoming would be in the case of curved lanes, this pipeline may not work as a left lane point may be classified as a right lane point as the segregation is based on slopes.

### 3. Suggest possible improvements to your pipeline

A possible improvement would be to incorporate the use of history, that is to see the old lane lines and decide the new ones if no lanes are detected due to various lightinf anomalies. 

Another potential improvement could be to apply clustering algorithms for seperation of lane points as left and right lane points rather than using the slope as a metric as I believe that is how our mind segregates the points and i am sure that can provide better results
